The RecursiveCharacterTextSplitter in LangChain is a text splitting mechanism designed to break down large documents into smaller, more manageable chunks while attempting to preserve semantic meaning and contextual continuity. It achieves this by recursively trying to split the text using a list of separators, moving to finer-grained separators if the previous split results in chunks that are still too large.

> Add blockquote


**How it works:**
***Prioritized Separators:*** It uses a list of characters as separators, by default ["\n\n", "\n", " ", ""], which prioritize splitting by paragraphs, then newlines, then spaces, and finally individual characters if necessary.

> Add blockquote


**Recursive Splitting:**
It first attempts to split the entire text by the first separator (e.g., \n\n for paragraphs).
For each resulting chunk, it checks if its size exceeds the specified chunk_size.
If a chunk is still too large, it recursively applies the next separator in the list (e.g., \n for newlines) to that chunk.
This process continues until all chunks are smaller than or equal to the chunk_size, or the finest-grained separator (individual characters) has been used.


> Add blockquote


**Chunk Size and Overlap:**
chunk_size: Defines the maximum size of each resulting text chunk.
chunk_overlap: Specifies the number of characters that will overlap between consecutive chunks. This helps maintain context when information might be split across chunk boundaries.


> Add blockquote


**Key Features:**
Semantic Preservation: By prioritizing larger, more meaningful separators (like paragraphs) first, it aims to keep related information together within chunks.


> Add blockquote


Flexibility: You can customize the list of separators and the chunk_size and chunk_overlap parameters to suit your specific document structure and downstream application needs.


> Add blockquote


Recursive Nature: The recursive application of separators ensures that even very long and complex documents can be effectively broken down.

```python
from langchain_huggingface import HuggingFacePipeline
from transformers import pipeline

# First, create a pipeline
pipe = pipeline(
    "text-generation",
    model="google/gemma-2b-it",
    temperature=0.2,
    max_new_tokens=512,
)

# Then pass it to HuggingFacePipeline
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableParallel, RunnablePassthrough
#from langchain_huggingface import HuggingFaceEndpoint

retriever = vectorstore.as_retriever()

system_prompt = (
    "You are an assistant for question-answering tasks. "
    "Use the following pieces of retrieved context to answer "
    "the question. If you don't know the answer, say that you "
    "don't know. Use three sentences maximum and keep the "
    "answer concise."
    "\n\n"
    "{context}"
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        ("human", "{input}"),
    ]
)

# Initialize LLM
# llm = HuggingFaceEndpoint(
#     repo_id="mistralai/Mistral-7B-Instruct-v0.2",
#     task="conversational",
#     huggingfacehub_api_token="your_api_token"
# )

# Build chains
question_answer_chain = prompt | llm | StrOutputParser()

rag_chain = (
    RunnableParallel(
        context=lambda x: "\n\n".join([doc.page_content for doc in retriever.invoke(x["input"])]),
        input=RunnablePassthrough()
    )
    | question_answer_chain
)


```

```python
result = rag_chain.invoke({"input": "What caused the oil spill in the San Francisco Bay on November 7, 2007?"})
print(result)
```
